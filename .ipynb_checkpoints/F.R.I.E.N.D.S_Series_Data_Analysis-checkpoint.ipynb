# Data Scraping from Wikipedia

Data scraping, also known as web scraping, is the process of importing information from a website into a spreadsheet or local file saved on your computer. It's one of the most efficient ways to get data from the web, and in some cases to channel that data to another website.

```python
# get HTML for wikipedia
url = "https://en.wikipedia.org/wiki/List_of_Friends_episodes"
uclient = ureq(url)
page_html = uclient.read()
uclient.close()

# get episode wise data on page
page_soup = soup(page_html,"html.parser")
seasons = page_soup.findAll("table",{"class":"wikitable plainrowheaders wikiepisodetable"})

# store data in pandas dataframe

column_names = ['season_no','episode_no','prod_code','episode_title','directed_by','written_by','orginal_air_date','us_viewers','network']
wiki_friends = pd.DataFrame(columns=column_names)

for season in seasons:
  episodes = season.findAll("tr",{"class":"vevent"})
  for episode in episodes:
    wiki_friends = wiki_friends.append({'episode_no': episode.findAll('td')[0].text, 'prod_code': episode.findAll('td')[5].text, 'episode_title': episode.findAll('td')[1].text, 'directed_by': episode.findAll('td')[2].text, 'written_by': episode.findAll('td')[3].text, 'orginal_air_date': episode.findAll('td')[4].span.span.text, 'us_viewers': episode.findAll('td')[6].text}, ignore_index=True)

# clean columns episode_title , us_viewers , season_no and written_by
wiki_friends['episode_title'] = wiki_friends['episode_title'].str.split('[').str[0].str.replace('"','')
wiki_friends['us_viewers'] = wiki_friends['us_viewers'].str.split('[').str[0]
wiki_friends['season_no'] = wiki_friends['prod_code'].str[:3]
wiki_friends['written_by'] = wiki_friends['written_by'].str.split(':').str[-1]

# adding value for column network
wiki_friends['network'] = 'NBC'


# cleaning of column season_no by replacing values
wiki_friends['season_no'].replace({'456':'1',
                                   '457':'2',
                                   '465':'3',
                                   '466':'4',
                                   '467':'5',
                                   '225':'6',
                                   '226':'7',
                                   '227':'8',
                                   '175':'9',
                                   '176':'10',
                                  },inplace=True)
```

### Data Visualizations : Exploring "Friends" series data for Wikipedia 

```python
# episode count by season
query=result.groupby(['season_no']).count().reset_index().sort_values(by='episode_no',ascending=False).head(10)


# Barplot using Seaborn
plt.figure(figsize=(20,7))

splot = sns.barplot(x="season_no",
            y="episode_no",
            data=query
            )

# Display values on top of Bars
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', 
                   va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points'
                  )
    
# Setting Bar Plot configuration
plt.xlabel('season_no', fontsize=20)
plt.ylabel('episode_no', fontsize=20)
plt.title('Episode Count By Season', fontsize=25)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()

# Crosstab - Table format data

query[['season_no','episode_no']]

# Seasons 1 to 9 have more than 24 episodes where as Season 10 has only 18 episodes.
```

```python
# episode count by writer

query=result.groupby(['written_by']).count().reset_index().sort_values(by='episode_no',ascending=False).head(10)


# Barplot using Seaborn
plt.figure(figsize=(20,7))

splot = sns.barplot(x="written_by",
            y="episode_no",
            data=query
            )

# Display values on top of Bars
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', 
                   va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points'
                  )
    
# Setting Bar Plot configuration
plt.xlabel('written_by', fontsize=20)
plt.ylabel('episode_no', fontsize=20)
plt.title('Episode Count By Top 10 Writers', fontsize=25)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15,horizontalalignment='right',rotation=45)
plt.show()

# Crosstab - Table format data

query[['written_by','episode_no']]

# The writers "Andrew Reich & Ted Cohen" wrote 5% (12/236) of the episodes across series.
# The top 10 writers combined wrote 37% (88/236) of the episodes across series.
```

```python
# episode count by Episode Type

query=result.groupby(['episode_type']).mean().reset_index().sort_values(by='us_viewers',ascending=False)


# Barplot using Seaborn
plt.figure(figsize=(17,7))

splot = sns.barplot(x="episode_type",
                    y="us_viewers",
                    data=query
                   )

# Display values on top of Bars
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', 
                   va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points'
                  )

# Setting Bar Plot configuration
plt.xlabel('episode_type', fontsize=20)
plt.ylabel('us_viewers', fontsize=20)
plt.title('Average US Viewers by Episode Type', fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15,horizontalalignment='right',rotation=45)
plt.show()

# Crosstab - Table format data

query[['episode_type','us_viewers']]

# Observations as listed below:
# Season Finale episodes have 31.77% more average US viewers than that of normal episodes.
# Season Premiere episodes have 15.56% more average US viewers than of normal episodes. 
# The 99th and 100th episodes had approximately same number of viewership.
```

```python
# total US viewers by season and episode

linequery=result.groupby(['prod_code']).mean().reset_index()

crosstab_query=result.groupby(['prod_code']).mean().reset_index().sort_values(by='us_viewers',ascending=False)

# Lineplot using Seaborn
plt.figure(figsize=(45,10))

sns.lineplot(x="prod_code",
             y="us_viewers",
             data=linequery,
             marker='o', 
             markersize=7,
             color='red',
             linewidth=1
            )

# Setting Line Plot configuration
plt.xlabel('prod_code', fontsize=25)
plt.ylabel('us_viewers', fontsize=25)
plt.title('Line Graph - Total US Viewers by Season & Episode', fontsize=30)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20,horizontalalignment='right',rotation=45)
plt.show()


# Crosstab - Table format data

crosstab_query[['prod_code', 'season_no', 'episode_no', 'us_viewers']]

# US viewers shows a mix trend with spikes for certain episodes. 
# Season 2 - Episode 12 and 13 and Season 10 - Episode 17 & 18 --> Most watched episodes
```

```python
# Average US viewers by season

linequery=result.groupby(['season_no']).mean().reset_index()

crosstab_query=result.groupby(['season_no']).mean().reset_index().sort_values(by='us_viewers',ascending=False)

# Lineplot using Seaborn
plt.figure(figsize=(15,5))

sns.lineplot(x="season_no",
            y="us_viewers",
            data=linequery,
            marker='o', 
            markersize=10,
            color='mediumvioletred',
            linewidth=2
            )

# Setting Line Plot configuration
plt.xlabel('season_no', fontsize=20)
plt.ylabel('us_viewers', fontsize=20)
plt.title('Line Graph - Average US Viewers by Season', fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()

# Crosstab - Table format data

crosstab_query[['season_no','us_viewers']]

# Observations as listed below:
# average US viewers decreases from Season 2 to Season 7 (30.64% decrease) and; 
# then increases 1.22 times (21% increase) in season 8 compared to season 7 and; 
# then decreases from Season 8 to Season 9 (10% decrease) and;
# then increases 1.1 times (9% increase) in Season 10 compared with Season 9.
# The Season 2 was the most popular across the series
```
