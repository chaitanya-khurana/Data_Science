```python
# to handle  data retrieval
import urllib3
from urllib3 import request
# to handle certificate verification
import certifi
# to manage json data
import json
# for pandas dataframes
import pandas as pd
```

```python
# handle certificate verification and SSL warnings
# https://urllib3.readthedocs.io/en/latest/user-guide.html#ssl
http = urllib3.PoolManager(
       cert_reqs='CERT_REQUIRED',
       ca_certs=certifi.where())
```

```python
# get data from the API
url = 'https://www.mldata.io/get-data/dataset/label-encoded/heart_disease/?format=json'
r = http.request('GET', url)
r.status
```

```python
# decode json data into a dict object
data = json.loads(r.data.decode('utf-8'))
data
```

```python
# get column headers from field attribute_names
column_names = data['attribute_names']
column_names
```

```python
# access values
value_access = data['values']
value_access
```

```python
# pull first 13 column values
valuesX = value_access['X']
valuesX
```

```python
# create a data frame with 13 column values
output = pd.DataFrame(data=valuesX)
```

```python
# adding column headers from "attribute names"
output.columns = column_names[:-1]
```

```python
# checking output
output
```

```python
# pull last column value for "num" field
valuesY = value_access['Y']
valuesY
```

```python
# add new column "num" to the output dataframe
output['num'] = valuesY
```

```python
# checking output
output
```

```python
# correct column datatypes for all fields

convert_dict = {'age': int, 
                'sex': int, 
                'cp': int, 
                'trestbps': int, 
                'chol': int, 
                'fbs': int, 
                'restecg': int, 
                'thalach': int, 
                'exang': int, 
                'oldpeak': float, 
                'slope': int, 
                'ca': int,
                'thal': int, 
                'num': int
               } 
  
output = output.astype(convert_dict) 
print(output.dtypes) 
```
